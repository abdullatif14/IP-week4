```
# checking for null values 

df.isnull()

#checking if there is any missing data in each column

df.isnull().any()

# checking how many missing values in each column

df.isnull().sum()

# we can do a quick check if we have any missing values at all

print(df.isnull().values.any())
```

```
# removing missing values 
# dropping the missing values

clean_df = df.dropna()
print(clean_df)

# checking if all missing values are dropped
print(clean_df.isnull().sum())
```

```
# dropping all NA values 
all_cols_any_na = df.dropna(axis=1)
print(all_cols_any_na)

# dropping all attributes that have NA values
all_cols_all_na = df.dropna(axis=1, how="all")
print(all_cols_all_na)
```
